```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('Malayalam_Char_Gabor.csv')

# Choose the two features to plot
x = data['Theta0_Lambda1_LocalEnergy']
y = data['Theta0_Lambda1_MeanAmplitude']

# Create a scatter plot
plt.scatter(x, y)
plt.xlabel('Theta0_Lambda1_LocalEnergy')
plt.ylabel('Theta0_Lambda1_MeanAmplitude')
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = data['Theta0_Lambda1_LocalEnergy'].values.reshape(-1, 1)
y = data['Theta0_Lambda1_MeanAmplitude'].values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print('Mean Square Error:', mse)
```

```python
model = LinearRegression()
model.fit(X, y)
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print('Mean Square Error:', mse)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load the data from the CSV file
data = pd.read_csv('Malayalam_Char_Gabor.csv')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.drop('Label', axis=1), data['Label'], test_size=0.2, random_state=42)

# Create a Regression Tree model
tree_model = DecisionTreeRegressor()
tree_model.fit(X_train, y_train)

# Create a k-NN Regressor model
knn_model = KNeighborsRegressor()
knn_model.fit(X_train, y_train)

# Predict the values using the models on the testing set
tree_pred = tree_model.predict(X_test)
knn_pred = knn_model.predict(X_test)

# Evaluate the performance of the models using MSE and R2 metrics
tree_mse = mean_squared_error(y_test, tree_pred)
tree_r2 = r2_score(y_test, tree_pred)
knn_mse = mean_squared_error(y_test, knn_pred)
knn_r2 = r2_score(y_test, knn_pred)

print('Regression Tree:')
print('MSE:', tree_mse)
print('R2:', tree_r2)

print('k-NN Regressor:')
print('MSE:', knn_mse)
print('R2:', knn_r2)
```
